```python
import subprocess
```

```python
script_str = '''from mpi4py import MPI

comm = MPI.COMM_WORLD
rank = comm.Get_rank()

color = rank % 2
local_comm = comm.Split(color)

if color == 0:

    from fitsnap3.parallel_tools import ParallelTools
    pt = ParallelTools(comm=local_comm)

    from fitsnap3.io.input import Config
    config = Config(arguments_lst = ["Ta-example.in"])

    from fitsnap3.fitsnap import FitSnap
    snap = FitSnap()
    snap.scrape_configs()
    snap.process_configs()
    pt.all_barrier()
    snap.perform_fit()
    snap.write_output()


local_rank = local_comm.Get_rank()
local_size = local_comm.Get_size()
print(rank, color, local_rank, local_size)'''
```

```python
with open('demompi.py', 'w') as f:
    f.writelines(script_str)
```

```python
subprocess.check_output('mpiexec --oversubscribe -n 4 python demompi.py', shell=True, universal_newlines=True ).split("\n")
```
